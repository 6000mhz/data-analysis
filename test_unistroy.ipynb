```python
# Данные взяты из открытых источноков ФНС https://www.nalog.gov.ru/opendata/

from zipfile import ZipFile

with ZipFile('C:/Users/ayrat/Downloads/data-20210801-structure-20180801.zip', 'r') as zipObj:
    zipObj.extractall('C:/Users/ayrat/Downloads/FNS_xml')
```

```python
import os
import xml.etree.ElementTree as Xet

def parse_dii():
    path = 'C:/Users/ayrat/Downloads/FNS_xml/'
    fileList = os.listdir(path)
    for t in fileList:
        with open('C:/Users/ayrat/Downloads/FNS_xml/' + t, 'r', encoding='utf-8') as ft:
            base = os.path.splitext(t)[0]
            with open(('C:/Users/ayrat/Downloads/FNS_csv/' + base + '.csv'), 'w+', encoding='utf-8') as f:
                xmlparse = Xet.parse('C:/Users/ayrat/Downloads/FNS_xml/' + t)
                root = xmlparse.getroot()
                for i in root:
                    for i2 in i.findall('СведНП'):
                        f.writelines(i2.attrib['НаимОрг'])
                        f.write(',')
                        f.writelines(i2.attrib['ИННЮЛ'])
                        f.write(',')
      
                        for i2 in i.findall('СведДохРасх'): 
                           f.writelines(i2.attrib['СумДоход'])
                           f.write(',')
                           f.writelines(i2.attrib['СумРасход'])
                           f.write('\n')

if __name__ == '__main__':
    parse_dii()
```

```python
import os
import xml.etree.ElementTree as Xet

def parse_dii():
    path = 'C:/Users/ayrat/Downloads/OOO_xml/'
    fileList = os.listdir(path)
    for t in fileList:
        with open('C:/Users/ayrat/Downloads/OOO_xml/' + t, 'r', encoding='utf-8') as ft:
            base = os.path.splitext(t)[0]
            with open(('C:/Users/ayrat/Downloads/OOO_csv/' + base + '.csv'), 'w+', encoding='utf-8') as f:
                xmlparse = Xet.parse('C:/Users/ayrat/Downloads/OOO_xml/' + t)
                root = xmlparse.getroot()
                for i in root:
                    for i2 in i.findall('СведНП'):
                        f.writelines(i2.attrib['НаимОрг'])
                        f.write(',')
                        f.writelines(i2.attrib['ИННЮЛ'])
                        f.write(',')
      
                        for i2 in i.findall('СведНаруш'): 
                           f.writelines(i2.attrib['СумШтраф'])
                           f.write('\n')

if __name__ == '__main__':
    parse_dii()
```

```python
OOO_1 = OOO.merge(FNS, left_on='ИННЮЛ', right_on='ИННЮЛ', how='outer')

OOO_1
```

```python
# Компании имеющие не нулевую прибыль безопасны
OOO_safe = OOO_1[OOO_1['Прибыль-Штраф'] > 0]

OOO_safe
```
